```python
import re
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as uReq
```

```python
my_url = 'https://www.newegg.com/Product/ProductList.aspx?Submit=ENE&N=100006662&IsNodeId=1&Description=graphics&name=Video%20Cards%20%26%20Video%20Devices&Order=BESTMATCH'

uClient = uReq(my_url)
page_html = uClient.read()
uClient.close()

with open('products.csv', 'w') as file:
    
    headers = 'Brand, Product-Name, Shipping-Charge\n'
    
    file.write(headers)
    
    page_soup = soup(page_html, 'html.parser')
    containers = page_soup.find_all('div', class_='item-container ')
    try:
        
        for container in containers:

            brand = container.div.div.img['title']

            title_container = container.find_all('a', class_= 'item-title')
            peoduct_name = title_container[0].text

            shipping_container = container.find_all('li', class_= 'price-ship')
            shipping = shipping_container[0].text.strip()
        

            print("brand : " + brand)
            print("peoduct_name : " + peoduct_name)
            print("shipping: " + shipping)
            
            file.write(brand +',' + peoduct_name.replace(',', '|') + ',' + shipping + '\n')
    except TypeError:
            print("Don't Care Carry On.")

    
    
```
